## Analisis

### Primera Parte

```python
# Cuantas selecciones participaron?
# Completar

Total_Selecciones = pd.concat([mundial['1'], mundial['2']]).nunique()
Total_Selecciones
```

```python
# Cuantas fases tiene el mundial?
# Completar

Fases = mundial['group'].nunique()
Fases
```

```python
# Cual fue la "attendance" promedio por partido?
# Completar

Attendance_Promedio = mundial['attendance'].mean()
Attendance_Promedio
```

```python
# Cual fue el partido con mas diferencia de goles?
# Completar

mundial['Diferencia_Goles'] = (mundial['1_goals'] - mundial['2_goals'])

Mayor_Diferencia = mundial['Diferencia_Goles'].max()

Partido_Mayor_Diferencia = mundial[mundial['Diferencia_Goles'] == Mayor_Diferencia]

Partido_Mayor_Diferencia
```

```python
# Cual es la selección que mas penales metió?
# Completar

Penales_Equipo1 = mundial.groupby('1')['1_panelties_scored'].sum()
Penales_Equipo2 = mundial.groupby('2')['2_panelties_scored'].sum()

Total_Penales = Penales_Equipo1 + Penales_Equipo2

Total_Penales
```

```python
# Cual es la selección que mas penales le hicieron?
# Completar

Penales_Equipo1 = mundial.groupby('1')['1_panelties_scored'].sum()
Penales_Equipo2 = mundial.groupby('2')['2_panelties_scored'].sum()

Total_Penales_Recibidos = Penales_Equipo1 - Penales_Equipo2

print("Selección con más penales:", Total_Penales_Recibidos.idxmin())

print("Cantidad de penales:", Total_Penales.max())

Total_Penales_Recibidos
```

### Segunda Parte

```python
equipo : str = "ARGENTINA"
```

```python
# Seleccionar todos los partidos que jugo el equipo de "local"
# Completar

filtrolocal : pd.Series = mundial['1'] == equipo
Partidos_Local = mundial[filtrolocal]
Partidos_Local
```

```python
# Seleccionar todos los partidos que jugo el equipo de "visitante"
# Completar

filtroVisitante : pd.Series = mundial['2'] == equipo
Partidos_Visitante = mundial[filtroVisitante]
Partidos_Visitante
```

```python
# Seleccionar todos los partidos que jugo el equipo (de "local" o "visitante")
# Completar

partidos : pd.DataFrame = pd.concat([Partidos_Local, Partidos_Visitante])
partidos
```

```python
# Cuantos goles hizo el equipo en total (obtener los goles metidos de local y sumarlos con los goles metidos de visitante)
# Completar

goles : int = Partidos_Local['1_goals'].sum() + Partidos_Visitante['2_goals'].sum()
goles
```

```python
# Cuantos goles hizo el equipo en promedio
# Completar

Total_Partidos_Equipo : int = len(partidos)
Promedio_Goles = goles / Total_Partidos_Equipo
Promedio_Goles
```

```python
# Cuantos partidos gano el equipo
# Completar

ganados : int = ((Partidos_Local['1_goals'] > Partidos_Local['2_goals']).sum() + (Partidos_Visitante['2_goals'] > Partidos_Visitante['1_goals']).sum())
ganados
```

```python
# Graficar los goles hechos por el equipo en cada partido

# Graficar goles por partido

goles_pp = [r['1_goals'] if r['1'] == equipo else r['2_goals'] for _, r in partidos.iterrows()]
rivales  = [f"vs {r['2']}" if r['1'] == equipo else f"vs {r['1']}" for _, r in partidos.iterrows()]

fig, ax = plt.subplots(figsize=(10, 4))
ax.bar(rivales, goles_pp, color='lightblue', edgecolor='steelblue')
ax.set_title(f"Goles de {equipo} por partido")
ax.set_ylabel("Goles")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Tercera Parte

```python
# Contar cuantos goles hizo cada equipo en total en el mundial
# Completar

goles_totales = mundial.groupby('1')['1_goals'].sum().add(
                mundial.groupby('2')['2_goals'].sum(), fill_value=0
                ).sort_values(ascending=False)
goles_totales
```

```python
# Graficar los goles por equipo en todo el mundial

# Completar

fig, ax = plt.subplots(figsize=(14, 5))
goles_totales.plot(kind='bar', ax=ax, color='steelblue')
ax.set_title("Goles por equipo - Qatar 2022")
ax.set_ylabel("Goles")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Cual es el equipo que hizo mas goles en el mundial?
# Completar

print(goles_totales.idxmax(), goles_totales.max())
```

```python
# Por cada partido, cual fue la diferencia de goles?
# Completar

mundial['diferencia'] = abs(mundial['1_goals'] - mundial['2_goals'])
mundial[['match_no', '1', '2', '1_goals', '2_goals', 'diferencia']]
```

```python
# Por cada partido, cual fue el equipo que gano?
# Completar

def ganador(r):
    if r['1_goals'] > r['2_goals']: return r['1']
    if r['2_goals'] > r['1_goals']: return r['2']
    return 'EMPATE'

mundial['ganador'] = mundial.apply(ganador, axis=1)
mundial[['match_no', '1', '2', '1_goals', '2_goals', 'ganador']]
```

```python
# En cada fase del mundial, cual fue el equipo que hizo mas goles?
# Completar

for fase, df in mundial.groupby('group'):
    g = df.groupby('1')['1_goals'].sum().add(df.groupby('2')['2_goals'].sum(), fill_value=0)
    print(f"{fase}: {g.idxmax()} ({int(g.max())} goles)")
```
