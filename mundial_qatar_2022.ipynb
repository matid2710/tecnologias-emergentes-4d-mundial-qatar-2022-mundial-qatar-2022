## Analisis

### Primera Parte

```python
# Cuantas selecciones participaron?
# Completar

Total_Selecciones = pd.concat([mundial['1'], mundial['2']]).nunique()
Total_Selecciones
```

```python
# Cuantas fases tiene el mundial?
# Completar

Fases = mundial['group'].nunique()
Fases
```

```python
# Cual fue la "attendance" promedio por partido?
# Completar

Attendance_Promedio = mundial['attendance'].mean()
Attendance_Promedio
```

```python
# Cual fue el partido con mas diferencia de goles?
# Completar

mundial['Diferencia_Goles'] = (mundial['1_goals'] - mundial['2_goals'])

Mayor_Diferencia = mundial['Diferencia_Goles'].max()

Partido_Mayor_Diferencia = mundial[mundial['Diferencia_Goles'] == Mayor_Diferencia]

Partido_Mayor_Diferencia
```

```python
# Cual es la selección que mas penales metió?
# Completar

Penales_Equipo1 = mundial.groupby('1')['1_panelties_scored'].sum()
Penales_Equipo2 = mundial.groupby('2')['2_panelties_scored'].sum()

Total_Penales = Penales_Equipo1 + Penales_Equipo2

Total_Penales
```

```python
# Cual es la selección que mas penales le hicieron?
# Completar

Penales_Equipo1 = mundial.groupby('1')['1_panelties_scored'].sum()
Penales_Equipo2 = mundial.groupby('2')['2_panelties_scored'].sum()

Total_Penales_Recibidos = Penales_Equipo1 - Penales_Equipo2

print("Selección con más penales:", Total_Penales_Recibidos.idxmin())

print("Cantidad de penales:", Total_Penales.min())

Total_Penales_Recibidos
```

### Segunda Parte

```python
equipo : str = "ARGENTINA"
```

```python
# Seleccionar todos los partidos que jugo el equipo de "local"
# Completar

filtrolocal : pd.Series = mundial['1'] == equipo
Partidos_Local = mundial[filtrolocal]
Partidos_Local
```

```python
# Seleccionar todos los partidos que jugo el equipo de "visitante"
# Completar

filtroVisitante : pd.Series = mundial['2'] == equipo
Partidos_Visitante = mundial[filtroVisitante]
Partidos_Visitante
```

```python
# Seleccionar todos los partidos que jugo el equipo (de "local" o "visitante")
# Completar

partidos : pd.DataFrame = pd.concat([Partidos_Local, Partidos_Visitante])
partidos
```

```python
# Cuantos goles hizo el equipo en total (obtener los goles metidos de local y sumarlos con los goles metidos de visitante)
# Completar

goles : int = Partidos_Local['1_goals'].sum() + Partidos_Visitante['2_goals'].sum()
goles
```

```python
# Cuantos goles hizo el equipo en promedio
# Completar

Total_Partidos_Equipo : int = len(partidos)
Promedio_Goles = goles / Total_Partidos_Equipo
Promedio_Goles
```

```python
# Cuantos partidos gano el equipo
# Completar

ganados : int = ((Partidos_Local['1_goals'] > Partidos_Local['2_goals']).sum() + (Partidos_Visitante['2_goals'] > Partidos_Visitante['1_goals']).sum())
ganados
```

```python
# Graficar los goles por equipo en todo el mundial

fig, ax = plt.subplots()
# Completar

```
